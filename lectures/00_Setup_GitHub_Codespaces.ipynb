## 1 · Get a GitHub account and fork the repository

*Skip step 1 if you already have a GitHub account.*

1. Create a free account at [github.com/signup](https://github.com/signup). Your username will be visible to the class — pick something you are happy to keep.
2. Go to the course repository page (the link was sent by the organizers, it looks like `github.com/laserlab/CountingPhotons`).
3. Click **Fork** (top right) → **Create fork**. This gives you *your own copy* of the course material under `github.com/<your-username>/CountingPhotons`. You can edit your fork freely without touching the original.

**Why a fork?** During the school you will complete exercises and push your solutions. You can only push to repositories you own — your fork is exactly that.

## 2 · Launch a GitHub Codespace (recommended way to run everything)

A *Codespace* is a Linux machine in the cloud, pre-configured by this repository's `.devcontainer` file, with VS Code running in your browser. Nothing gets installed on your laptop.

1. On **your fork's** page: green **Code** button → **Codespaces** tab → **Create codespace on main**.
2. Wait for the container build (~2 min the first time; afterwards it starts in seconds).
3. In the file explorer on the left, open `lectures/00_Setup_GitHub_Codespaces.ipynb` (this notebook!).
4. Click any code cell, press **Shift+Enter**. When asked for a kernel, choose **Python 3**.

> **Free tier:** GitHub gives every account 120 core-hours of Codespaces per month — far more than this school needs. Codespaces auto-sleep after 30 min of inactivity; your files survive. **Stop** (don't delete!) your codespace when done: Code → Codespaces → ⋯ → *Stop codespace*.

## 3 · Alternative: Google Colab

If Codespaces doesn't work for you, [Google Colab](https://colab.research.google.com) runs the same notebooks:

1. Go to [colab.research.google.com](https://colab.research.google.com) → **File ▸ Open notebook ▸ GitHub**.
2. Paste the URL of *your fork* and pick a notebook.
3. Run the first code cell — on Colab it installs QuTiP and Perceval automatically (takes ~1 min).

To save work back to GitHub from Colab: **File ▸ Save a copy in GitHub** and choose your fork (optional — nothing needs to be handed in).

````python
# Run me! (Shift+Enter)
# On Codespaces / a local install this cell does nothing (packages are already
# there). On Google Colab it installs the two packages the course needs.
import importlib.util
import subprocess
import sys

for module, pip_name in [("qutip", "qutip"),
                         ("perceval", "perceval-quandela")]:
    if importlib.util.find_spec(module) is None:
        print(f"installing {pip_name} ...")
        subprocess.check_call([sys.executable, "-m", "pip", "install",
                               "--quiet", pip_name])
print("environment ready")
````

## 5 · Troubleshooting

| Symptom | Fix |
|---|---|
| *"Select kernel"* dropdown shows nothing | Reload the browser tab; if it persists: kernel picker → *Python Environments* → `/usr/local/bin/python` |
| Codespace build fails / hangs | Delete the codespace (Code → Codespaces → ⋯ → Delete) and create a new one |
| `ModuleNotFoundError: qutip` in Codespaces | Open a terminal and run `pip install -r requirements/requirements.txt` |
| Colab: `pip install` takes forever | Normal on first run (~1 min); Runtime ▸ Restart session afterwards if imports still fail |
| `git push` asks for a password | In Codespaces this never happens (already authenticated). Locally: use a [personal access token](https://docs.github.com/en/authentication/keeping-your-account-and-data-secure/managing-your-personal-access-tokens) or the `gh auth login` CLI |
| Anything else | Come early on Monday, or open an issue on the course repository |
